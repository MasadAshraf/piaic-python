```python
user_monthly_units = 400


def electricity_cal(units):
    charges = 0
    if units > 0 and units <= 100:
        charges = units * 9.43
    elif units > 100 and units <= 200:
        charges = units * 10.29
    elif units > 200 and units <= 300:
        charges = units * 12.55
    elif units > 300 and units <= 400:
        charges = units * 17.38
    elif units > 400 and units <= 500:
         charges = units * 18.84
    elif units > 500 and units <= 600:
        charges = units * 19.76
    elif units > 600 and units <= 700:
        charges = units * 20.40
    elif units > 700:
        charges = units * 23.87
    else:
        print('invalid input')
        charges = -1

    return charges


def uniform_cal(units):
    return units * 53

def fuel_cal(units):
    return units * 3

def sales_tax_cal(units):
    return units * 3.4

def advance_tax_cal(units):
    return units * 2.2

def tv_fee_cal(units):
    return units * 0.2

def total_cal(e_charges, u_charges, f_charges ,s_tax ,a_tax ,t_fee):
    return e_charges + u_charges + f_charges + s_tax + a_tax + t_fee 



def bill_main(monthly_units):
    
    if monthly_units < 0:
        print('\033[91m' + "INVALID INPUT")
        return
    
    electric_charges= electricity_cal(monthly_units)
    print("electric charges for ", monthly_units, "=", electric_charges)
    
    uniform_charges= uniform_cal(monthly_units)
    print("uniform charges for ", monthly_units, "=", uniform_charges)
    
    fuel_charges= fuel_cal(monthly_units)
    print("fuel charges for ", monthly_units, "=", fuel_charges)
    
    sales_tax= sales_tax_cal(monthly_units)
    print("electric charges for ", monthly_units, "=", sales_tax)
    
    advance_tax= advance_tax_cal(monthly_units)
    print("electric charges for ", monthly_units, "=", advance_tax)
    
    tv_fee= tv_fee_cal(monthly_units)
    print("electric charges for ", monthly_units, "=", tv_fee)
    
    total= total_cal(electric_charges , uniform_charges , fuel_charges , sales_tax , advance_tax , tv_fee)
    print("#################################################")
    print("Total charges for ", monthly_units, "=", total)
    print("#################################################")
    

bill_main(user_monthly_units)
```
